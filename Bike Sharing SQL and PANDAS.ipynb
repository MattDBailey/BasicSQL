```python
import sqlite3
```

```python
import pandas as pd

db = sqlite3.connect('hubway.db')
res = db.execute("SELECT name FROM sqlite_master WHERE type='table';")
for name in res:
    print (name[0])

def run_query(query):
    return pd.read_sql_query(query,db)
```

```python
query = 'SELECT * FROM trips LIMIT 5;'
run_query(query)
```

```python
query = 'SELECT duration, start_date FROM trips LIMIT 10'
run_query(query)
```

```python
query = 'SELECT duration, start_date FROM trips ORDER BY duration DESC LIMIT 10'
run_query(query)
```

```python
query = '''
SELECT stations.station AS "Station", COUNT(*) AS "Count"
FROM trips
INNER JOIN stations
ON trips.start_station = stations.id
GROUP BY stations.station
ORDER BY COUNT(*) DESC
LIMIT 5;
'''

run_query(query)
```

```python
query = '''
SELECT stations.station AS "Station", COUNT(*) AS "Count"
FROM trips
INNER JOIN stations
ON trips.start_station = stations.id
WHERE trips.start_station = trips.end_station
GROUP BY stations.station
ORDER BY COUNT(*) DESC
LIMIT 5;
'''

run_query(query)
```

```python
query = '''
SELECT stations.municipality AS "Muni", AVG(duration) AS "Avg Time"
FROM trips 
INNER JOIN stations
ON trips.start_station = stations.id
GROUP BY stations.municipality
ORDER BY AVG(duration) DESC
LIMIT 5;
'''

run_query(query)
```

How many trips incurred additional fees (lasted longer than 30 minutes)?

```python
query = '''
SELECT COUNT(*) as "count"
FROM trips
WHERE duration > 1800;
'''

run_query(query)
```

Which bike was used for the longest total time?

```python
query = '''
SELECT bike_number, SUM(duration) as "Total Time"
FROM trips
GROUP BY bike_number
ORDER BY SUM(duration) DESC 
LIMIT 10;'''

run_query(query)
```

Did registered or casual users take more round trips?

```python
query = '''
SELECT sub_type, COUNT(*) AS "Count"
FROM trips
WHERE trips.start_station = trips.end_station
GROUP BY sub_type;
'''

run_query(query)
```

Which municipality had the longest average duration?

```python
query = '''
SELECT stations.municipality, AVG(trips.duration)/60 as "Avg Trip Time"
FROM trips
INNER JOIN stations
ON trips.start_station = stations.id
WHERE trips.start_station = trips.end_station
GROUP BY stations.municipality
ORDER BY AVG(trips.duration) DESC;
'''

run_query(query)
```
